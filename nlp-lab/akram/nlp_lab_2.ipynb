```python
import requests
from bs4 import BeautifulSoup
import re

from nltk.corpus import words

from collections import defaultdict
```

```python
import nltk
nltk.download('words')
```

```python
def scrape_and_clean_text(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    title = soup.find("h1", {"id": "firstHeading"}).text.strip() if soup.find("h1", {"id": "firstHeading"}) else "NoTitleFound"

    paragraphs = soup.find_all("p")
    cleaned_text = []

    for para in paragraphs:
        text = para.get_text(strip=True)
        if text:
            cleaned_text.append(text.replace(" ", ""))
        if len(cleaned_text) >= 3:
            break

    article_text = "".join(cleaned_text)

    article_text = re.sub(r'http\S+|www\.\S+', '', article_text)

    article_text = re.sub(r'[^A-Za-z0-9,.!?]+', '', article_text)

    return f"Title:{title}\n\nText:{article_text}"
```

```python
url = 'https://www.geeksforgeeks.org/supervised-machine-learning/?ref=lbp'
string = scrape_and_clean_text(url)
```

## Maximum Matching Algorithm

```python
def segment_string(string):
    tokens = []
    lowercase_corpus = set([x.lower() for x in words.words()])
    i = 0

    while i < len(string):
        max_word = ""
        for j in range(i, len(string)):
            temp_word = string[i:j+1]
            if temp_word.lower() in lowercase_corpus and len(temp_word) > len(max_word):
                max_word = temp_word

        if max_word:
            tokens.append(max_word)
            i += len(max_word)
        else:
            match = re.match(r'[^A-Za-z0-9]+', string[i:])
            if match:
                tokens.append(match.group())
                i += len(match.group())
            else:
                tokens.append(string[i])  # Preserve non-word characters
                i += 1

    return tokens
```

## Byte-Pair Encoding

```python
def get_stats(vocab):
    pairs = defaultdict(int)
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols)-1):
            pairs[symbols[i],symbols[i+1]] += freq
    return pairs
```

```python
def merge_vocab(pair, v_in):
    v_out = {}
    bigram = re.escape(' '.join(pair))
    p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
    for word in v_in:
        w_out = p.sub(''.join(pair), word)
        v_out[w_out] = v_in[word]
    return v_out
```

```python
def get_vocab(data):
    vocab = defaultdict(int)
    for line in data:
        for word in line.split():
            vocab[' '.join(list(word)) + ' </w>'] += 1
    return vocab
```

```python
def byte_pair_encoding(data, n):
    vocab = get_vocab(data)
    for i in range(n):
        pairs = get_stats(vocab)
        print(pairs)
        best = max(pairs, key=pairs.get)
        vocab = merge_vocab(best, vocab)
    return vocab
```
