```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Download necessary NLTK data
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Function to get POS tag
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

lemmatizer = WordNetLemmatizer()
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
stemmed_data=[]
lemmatized_data=[]
```

```python
for i in range(len(docs)):
     stemmed_data.append(stemmer.stem(docs[i]))
        
```

```python
for i in range(len(docs)):
        lemmatized_data.append(lemmatizer.lemmatize(docs[i],get_wordnet_pos(docs[i])))
        
```

```python
import requests
from bs4 import BeautifulSoup

def get_stuff(l):
    link=l
    data=requests.get(link)
    soup_1 = BeautifulSoup(data.text, "html.parser")
    para=soup_1.find_all("p")
    docs=[]
    for p in para :
        docs.append(p.get_text())
    uni=[]
    s=" "
    for d in docs:
        s+=''.join(set(d))
    uni=list(set(s))
    special_char=[]
    for  i in uni :
        if not i.isalpha() :
            special_char.append(i)
    special_char.remove(" ")
    for d in range(len(docs)) :
        for i in special_char :
            docs[d]=docs[d].replace(i,"")
    token=[]
    for d in docs :
        token.append(d.split(" "))
    unique_words=[]
    for i in range(len(token)):
        unique_words.extend(list(set(token[i])))
    unique_words=list(set(unique_words))
    unique_words.remove("")
    words=[]
    for i in token :
        words.extend(i)
    print(f"website : {l}")
    print(f"Vocabulary : {len(unique_words)}")
    print(f"Total number of words : {len(words)}")
    print(f"Number of Documents : {len(docs)}")
    print("\n")


```

```python
links=["https://en.wikipedia.org/wiki/Machine_learning",
       "https://en.wikipedia.org/wiki/Natural_language_processing",
       "https://en.wikipedia.org/wiki/Deep_learning",
       "https://en.wikipedia.org/wiki/Artificial_intelligence"
      ]
```

```python
for l in links :
    get_stuff(l)
```
